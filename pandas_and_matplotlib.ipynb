```python
import pandas as pd
df=pd.read_csv("/content/simple_used_car_dataset.csv")
df['Price'].mean()
```

```python
df['Fuel_Type'].fillna(df['Fuel_Type'].mode()[0],inplace=True)
df['Kilometers_Driven'].fillna(df['Kilometers_Driven'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# to check duplicate values
df[df.duplicated()]
```

```python
#treatement of duplicates
df.drop_duplicates(subset='Year',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Fuel_Type']=le.fit_transform(df[['Fuel_Type']])
df
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
mm=MinMaxScaler()
df['Kilometers_Driven']=mm.fit_transform(df[['Kilometers_Driven']])
df
```
